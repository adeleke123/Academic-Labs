## Task 1

Suppose you take out a loan with a principal amount of #100,000. If the annual interest rate on the loan is 5% and you plan to repay it in 3 years, how much interest will you owe at the end of the loan term?

For this task write a function called `interest()` that accepts the 3 parameters as input arguments and calculates the amount of interest owed at the end of the loan term.
The 3 input arguments should be in the given order:
- The principle loan amount
- The annual interest rate
- The number of years in which the loan should be repayed

Once you have written the function, call it with the provided parameters:
- Principal (P) = #100,000
- Rate (R) = 5% per annum
- Time (T) = 3 years

## Solution



```python
### CG T1
### Write your function here:

def interest(p, r, t):
    # Calculate interest: P * R * T
    # Note: we divide r by 100 to convert the percentage to a decimal
    total_interest = p * (r / 100) * t
    return total_interest

# Calling the function with your specific parameters
principal = 100000
rate = 5
time = 3

result = interest(principal, rate, time)

print(f"The total interest owed is: {result}")

```

## Task 2

Write a function called `celsius_to_fahrenheit()` that converts a temperature from degrees Celsius to degrees Fahrenheit.

You can use this formula:

$T_c ∗ 1.8 + 32$

where $T_c$ is the temperature in degrees celsius to be converted.

Once you have written the function, use it to convert the following temperatures from Celsius to Fahrenheit:

1. 100

2. 75

3. 120

## Solution

```python
### CG T2
### Write your function here:


def celsius_to_fahrenheit(celsius):
    # Apply the formula: (Celsius * 1.8) + 32
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

# List of temperatures to convert
temps_to_convert = [100, 75, 120]

# Loop through and print the results
for temp in temps_to_convert:
    result = celsius_to_fahrenheit(temp)
    print(f"{temp}°C is equal to {result}°F")
```
